# EXTRACTION DATA FOR ARUNACHAL PRADESH

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


data = []
for i in range(1,4):
    url = r'https://results.eci.gov.in/AcResultGen2ndJune2024/statewiseS02'+str(i)+'.htm'
# Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Check for HTTP issues
    
    # Parse the HTML, ensuring the correct encoding is used
    soup = BeautifulSoup(response.content, 'html.parser', from_encoding=response.encoding)
    
    # Find the table containing the results
    table = soup.find('table', {'class': 'table-party'})  # Adjust the class name if necessary
    
    # Data extraction from the table
    
    for row in soup.find_all('tr')[1:]:  # Skip header row
        cols = row.find_all('td')
        if len(cols) >= 28 :  # Ensure all columns are present
            data.append({
                'Constituency': cols[0].text.strip(),
                'Const. No.': cols[1].text.strip(),
                'Leading Candidate': cols[2].text.strip(),
                'Leading Party': cols[3].text.strip(),
                'Trailing Candidate': cols[15].text.strip(),
                'Trailing Party': cols[16].text.strip(),
                'Margin': cols[28].text.strip(),
                'Round': cols[29].text.strip(),
                'Status': cols[30].text.strip()
            })

# Create a DataFrame
df = pd.DataFrame(data)
df.to_csv('election_results_arunachal.csv', index=False)

print('Data has been extracted and saved to election_results.csv')
```

# ANALYSIS  ARUNACHAL PRADESH

```python
import matplotlib.pyplot as plt
```

```python
arunachal = pd.read_csv('election_results_arunachal.csv')
```

```python
arunachal['Leading'] = arunachal['Leading Party'].str.split('iParty Wise State').str[0]
```

```python
arunachal['Trailing'] = arunachal['Trailing Party'].str.split('\niParty Wise State').str[0]
```

```python
arunachal
```

```python
seats_won_arunachal = arunachal.groupby('Leading').count()
```

```python
seats_won_arunachal
```

```python
seats_won_arunachal.plot(kind='bar', color='skyblue')
plt.title('Seats Won by Each Party in Arunachal Pradesh')
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.show()
```

```python
df=arunachal
df_won = df[df['Status'] == 'Result Declared']
seats_won = df_won['Leading'].value_counts().reset_index()
seats_won.columns = ['Party', 'Seats Won']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
wedges, texts = ax.pie(seats_won['Seats Won'], labels=seats_won['Party'], wedgeprops=dict(width=0.3), startangle=-40)
plt.setp(wedges, width=0.3, edgecolor='white')

centre_circle = plt.Circle((0, 0), 0.70, color='white', fc='white', linewidth=0)
fig.gca().add_artist(centre_circle)

ax.set_aspect('equal') 
plt.title('Party Wise Results')
plt.show()
```

```python
andra['Trailing'] = andra['Trailing Party'].str.split('\niParty Wise State').str[0]
```

```python
andra
```

```python
seats_won_andhra = andra.groupby('Leading').count()
```

```python
seats_won_andhra
```

```python
seats_won_andhra.plot(kind='bar', color='skyblue')
plt.title('Seats Won by Each Party in Andra Pradesh')
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.show()
```

```python
df=andra
df_won = df[df['Status'] == 'Result Declared']
seats_won = df_won['Leading'].value_counts().reset_index()
seats_won.columns = ['Party', 'Seats Won']
```

```python
import matplotlib.pyplot as plt

# Create doughnut chart
fig, ax = plt.subplots()
wedges, texts = ax.pie(seats_won['Seats Won'], labels=seats_won['Party'], wedgeprops=dict(width=0.3), startangle=-40)
plt.setp(wedges, width=0.3, edgecolor='white')

# Adding circle in the center
centre_circle = plt.Circle((0, 0), 0.70, color='white', fc='white', linewidth=0)
fig.gca().add_artist(centre_circle)

ax.set_aspect('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Party Wise Results')
plt.show()
```

```python

vote_shares = df.groupby('Leading')['Margin'].sum() / df['Margin'].sum() * 100
vote_shares = vote_shares.reset_index()
vote_shares.columns = ['Party', 'Vote Share']


fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(vote_shares['Vote Share'], labels=vote_shares['Party'], autopct='%1.1f%%', startangle=90)
ax.set_aspect('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Party Wise Vote Share')
plt.show()
```
